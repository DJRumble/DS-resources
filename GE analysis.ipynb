```python
import pandas as pd
```

```python
df_election = pd.read_csv('HoC-GE2019-by-constituency.csv')
df_ref = pd.read_csv('EU-referendum-result-data.csv')
```

```python
df_election.head()
```

```python
df_ref.head()
```

```python
#Agg on region
df_ref_agg_rgion = df_ref.groupby(['Region']).sum().reset_index()
df_ref_agg_rgion['Remain_pct'] = df_ref_agg_rgion['Remain']/df_ref_agg_rgion['Valid_Votes']
df_ref_agg_rgion['Leave_pct'] = df_ref_agg_rgion['Leave']/df_ref_agg_rgion['Valid_Votes']
df_ref_agg_rgion['region_name'] = df_ref_agg_rgion['Region']
df_ref_agg_rgion.head()
```

```python
#Agg on Area
df_ref_agg_area = df_ref.groupby(['Area']).sum().reset_index()
df_ref_agg_area['Remain_pct'] = df_ref_agg_area['Remain']/df_ref_agg_area['Valid_Votes']
df_ref_agg_area['Leave_pct'] = df_ref_agg_area['Leave']/df_ref_agg_area['Valid_Votes']
df_ref_agg_area['county_name'] = df_ref_agg_area['Area']
df_ref_agg_area.head()
```

```python
df_election_region_join = pd.merge(df_election,
                 df_ref_agg_rgion[['Leave_pct', 'Remain_pct','region_name']],
                 on='region_name', 
                 how='left')
```

```python
df_election_area_join = pd.merge(df_election,
                 df_ref_agg_area[['Leave_pct', 'Remain_pct','county_name']],
                 on='county_name', 
                 how='left')
```
